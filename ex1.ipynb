```python
from scipy import stats

data = [1,2,3,4,5,6,8,8,8,8,9,10,10,10,12,14,18,18,18,22,39,44,55,55,55,55,66,78,79,88]

print("\nDetails of the data")
print(stats.describe(data))

print("\nCumulative Frequency")
print(stats.cumfreq(data))

print("\nGeometric Mean")
print(stats.gmean(data))

print("\nHarmonic Mean")
print(stats.hmean(data))

print("\nInter Quartile Range")
print(stats.iqr(data))

print("\nZ Score")
print(stats.zscore(data))

print("\nSkewness")
print(stats.skew(data))

print("\nKurtosis")
print(stats.kurtosis(data))

# Correlation

data = [1,2,3,6,8,10]
data1 = [2,3,4,5,9,12]

print("\nSpearman Correlation")
print(stats.spearmanr(data, data1))
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing

col_list = ["id","first","last","gender","Marks","selected"]
df = pd.read_csv("sample.csv", usecols=col_list)

print("Original Dataset")
print(df)
print("\nEnd of Listing\n")

df_gender_encode = LabelEncoder()
df["gender"] = df_gender_encode.fit_transform(df["gender"])

print("After Label Encoding")
print(df)
print("\nEnd of Listing\n")

df["Marks"] = preprocessing.scale(df["Marks"])
scaled_df = preprocessing.scale(df["Marks"])

print("After Scaling")
print(df)
print("\nScaling completed\n")

newarr = scaled_df.reshape(-1,1)
scaled_df_bin = preprocessing.Binarizer(threshold=0.5).transform(newarr)

df["Marks"] = scaled_df_bin

print("After Binarization")
print(df)
print("\nBinarization completed")
```

```python
import pandas as pd

col_list = ["id","first","last","gender","Marks","selected"]
df = pd.read_csv("sample.csv", usecols=col_list)

print("Original Dataset")
print(df)

df_duplicated = pd.concat([df]*2, ignore_index=True)

print("\nDataset with Duplicates")
print(df_duplicated)

df_duplicates_removed = df_duplicated.drop_duplicates()

print("\nDataset after Removing Duplicates")
print(df_duplicates_removed)
```

```python
import pandas as pd

df = pd.DataFrame({
    'm1':[50,'A',60,'A',80],
    'm2':[60,'A','60','A',80],
    'm3':[50,70,'A','A',60],
    'm4':[60,'A','A','A',60],
    'm5':['A','A','A',10,20]
})

df = df.apply(pd.to_numeric, errors='coerce')

print("DataFrame with NaN")
print(df)

df['m5'] = df['m5'].fillna(0)

print("\nAfter fillna(0) for m5")
print(df)

df1 = df.copy()
df1['m2'] = df1['m2'].fillna(df1['m2'].mean())

print("\nAfter filling m2 with Mean")
print(df1)

df2 = df.copy()
df2['m3'] = df2['m3'].fillna(df2['m3'].median())

print("\nAfter filling m3 with Median")
print(df2)

df3 = df.dropna(axis=1)

print("\nAfter Dropping Columns with NaN")
print(df3)
```

```python
from numpy import asarray
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

data = asarray([
    [1,3],
    [8,5],
    [6,7],
    [8,9]
])

print("Original Data")
print(data)

scaler1 = MinMaxScaler()
scaled1 = scaler1.fit_transform(data)

scaler2 = StandardScaler()
scaled2 = scaler2.fit_transform(data)

print("\nMinMax Scaling")
print(scaled1)

print("\nStandard Scaling (Z-Score)")
print(scaled2)
```

```python
import pandas as pd
from matplotlib import pyplot as plt

# Read CSV file
col_list = ['id','name','gender','physics','chemistry','maths','biology','language','selected']
df = pd.read_csv('marks.csv', usecols=col_list)

print(df)


marks1 = df['physics'].values.tolist()

df['physics'].hist()

plt.title("Histogram for Physics Marks")
plt.xlabel("Physics Marks")
plt.ylabel("Frequency")
plt.show()


chemistry1 = df['chemistry'].values.tolist()

plt.scatter(chemistry1, marks1, alpha=0.5)

plt.title("Scatter Plot - Chemistry vs Physics")
plt.xlabel("Chemistry")
plt.ylabel("Physics")
plt.show()


df.plot.box(title="Box and Whisker Plot of Marks", grid=False)

plt.show()


sums = df.groupby("gender")["selected"].count()

plt.pie(sums, labels=sums.index, autopct='%1.1f%%')

plt.title("Students by Gender")

plt.show()


df.plot(
    kind='box',
    subplots=True,
    layout=(3,3),
    figsize=(10,8),
    sharex=False,
    sharey=False
)

plt.show()


df2 = df.drop(columns=['id','name','gender','selected'])

pd.plotting.scatter_matrix(df2, alpha=0.7, figsize=(10,10))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("iris.csv")

data = df.iloc[1:20]

print(data)

# Scatter Plot
data.plot.scatter(
    x='sepal.length',
    y='sepal.width',
    title="Sepal Length vs Sepal Width"
)

plt.show()

# Histogram
data.hist(
    column='sepal.length',
    by='variety',
    figsize=(8,5)
)

plt.show()

# Box Plot
data.boxplot(
    column='sepal.length',
    by='variety'
)

plt.show()
```
